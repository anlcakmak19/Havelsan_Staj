# K-fold cross validation

```python
# Define the number of folds
k = 5

# Randomly shuffle the indices of the data
indices = np.arange(X.shape[0])
np.random.shuffle(indices)

# Split the indices into k groups
indices = np.array_split(indices, k)

# Perform k-fold cross-validation
accuracies = []
```

```python
indices
```

```python
for i in range(k):
    # Prepare the test set indices
    test_indices = indices[i]
    
    # Prepare the training set indices
    train_indices = np.concatenate([indices[j] for j in range(k) if j != i])
    
    # Get the training and test sets
    X_train, X_test = X.iloc[train_indices], X.iloc[test_indices]
    y_train, y_test = y.iloc[train_indices], y.iloc[test_indices]
    # Initializing the Random Forest Regression model with 10 decision trees
    model = RandomForestRegressor(n_estimators = 10, random_state = 0)

    # Fitting the Random Forest Regression model to the data
    model.fit(X_train, y_train)
    # Test the model
    y_pred = model.predict(X_test)
    
    # Calculate the accuracy and append it to the list
    accuracy = np.sqrt(mean_squared_error(y_test, y_pred))
    accuracies.append(accuracy)
```

```python
# Print the average accuracy
print("Average accuracy: ", np.mean(accuracies))
```

# Randomized Search

```python
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_features = ['auto', 'sqrt','log2']
max_depth = [int(x) for x in np.linspace(10, 1000,10)]
min_samples_split = [2, 5, 10,14]
min_samples_leaf = [1, 2, 4,6,8]
```

```python
model = RandomForestRegressor()
best_score = np.inf
n_iterations = 10
# Splitting the dataset into training and testing set (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 28)
for i in range(n_iterations):
    start_time = time.time()
    # Select a random values for parameters
    n_estimator = random.choice(n_estimators)
    max_feature = random.choice(max_features)
    max_dep = random.choice(max_depth)
    min_samples_spl = random.choice(min_samples_split)
    min_samples_lf = random.choice(min_samples_leaf)
    # Set parameters to model
    model.set_params(n_estimators=n_estimator, max_features=max_feature, max_depth=max_dep, min_samples_split=min_samples_spl, min_samples_leaf=min_samples_lf)
    
    # Fit model on training data
    model.fit(X_train, y_train)
    paramaters_dict={'n_estimator':n_estimator,'max_feature':max_feature,'max_dep':max_dep,'min_samples_spl':min_samples_spl,'min_samples_lf':min_samples_lf}
    # Make predictions and get the error
    predictions = model.predict(X_test)
    error = np.sqrt(mean_squared_error(y_test, predictions))
    # If the score of the current iteration is better than best recorded so far, update best_score and best_params
    if error < best_score:
        best_score = error
        best_params = {"n_estimators": n_estimator, "max_features": max_feature, "max_depth": max_dep, "min_samples_split": min_samples_spl, "min_samples_leaf": min_samples_lf}
    end_time = time.time()
    iteration_time = end_time - start_time
    print(color.BOLD)
    print("------------------------------------------------------------------------------------------------------")
    print("Iteration {0} took {1} seconds.".format(i, iteration_time))
    print(paramaters_dict)
    print("root_mean_squared_error: ",error)
    print("------------------------------------------------------------------------------------------------------")
    print(color.END)
print("Best parameters found: ", best_params)
print("Lowest RMSE found: ",(best_score))
```

# Randomized Search with K-fold cross validation


```python
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_features = ['auto', 'sqrt','log2']
max_depth = [int(x) for x in np.linspace(10, 1000,10)]
min_samples_split = [2, 5, 10,14]
min_samples_leaf = [1, 2, 4,6,8]
```

```python
# Define your model 
model = RandomForestRegressor()

# Assuming X_train, y_train are already defined
best_score = np.inf

# Number of iterations for the search
n_iterations = 10
k = 5
# Randomly shuffle the indices of the data
indices = np.arange(X.shape[0])
np.random.shuffle(indices)

# Split the indices into k groups
indices = np.array_split(indices, k)

# Perform k-fold cross-validation
accuracies = []

for i in range(n_iterations):
    start_time = time.time()
    # Select a random values for parameters
    n_estimator = random.choice(n_estimators)
    max_feature = random.choice(max_features)
    max_dep = random.choice(max_depth)
    min_samples_spl = random.choice(min_samples_split)
    min_samples_lf = random.choice(min_samples_leaf)
    # Set parameters to model
    model.set_params(n_estimators=n_estimator, max_features=max_feature, max_depth=max_dep, min_samples_split=min_samples_spl, min_samples_leaf=min_samples_lf)
    paramaters_dict = {"n_estimators": n_estimator, "max_features": max_feature, "max_depth": max_dep, "min_samples_split": min_samples_spl, "min_samples_leaf": min_samples_lf}
    # Perform k-fold cross-validation
    accuracies = []
    for v in range(k):
        # Prepare the test set indices
        test_indices = indices[v]

        # Prepare the training set indices
        train_indices = np.concatenate([indices[j] for j in range(k) if j != v])

        # Get the training and test sets
        X_train, X_test = X.iloc[train_indices], X.iloc[test_indices]
        y_train, y_test = y.iloc[train_indices], y.iloc[test_indices]
        # Initializing the Random Forest Regression model with 10 decision trees
        # Fitting the Random Forest Regression model to the data
        model.fit(X_train, y_train)
        # Test the model
        y_pred = model.predict(X_test)

        # Calculate the accuracy and append it to the list
        accuracy = np.sqrt(mean_squared_error(y_test, y_pred))
        accuracies.append(accuracy)
    error = np.mean(accuracies)
    if error < best_score:
        best_score = error
        best_params = {"n_estimators": n_estimator, "max_features": max_feature, "max_depth": max_dep, "min_samples_split": min_samples_spl, "min_samples_leaf": min_samples_lf}
    end_time = time.time()
    iteration_time = end_time - start_time
    print(color.BOLD)
    print("------------------------------------------------------------------------------------------------------")
    print("Iteration {0} took {1} seconds.".format(i, iteration_time))
    print(paramaters_dict)
    print("root_mean_squared_error: ",error)
    print("------------------------------------------------------------------------------------------------------")
    print(color.END)
print("Best parameters found: ", best_params)
print("Lowest RMSE found: ",(best_score))
        
```
